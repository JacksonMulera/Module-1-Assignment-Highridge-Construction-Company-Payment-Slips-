```python
import random as rn
```

```python
employees = []
payslip_count=0
for i in range(1, 401):
    name = f"employee{i}"
    salary=rn.randint(7500,30001)
    gender=rn.choice(["Male","Female"])
    employee_level=""
    if 10000<salary<20000:
        employee_level="A1"
    elif 7500<salary<30001 and gender=="Female":
        employee_level="A5-F"
    else:
        employee_level="N/A"
    employees.append({"name": name, "salary": salary, "gender": gender,"employee_level":employee_level})
```

```python
# Print the employees
for employee in employees:
    print(f"Name: {employee['name']}")
    print(f"Salary: {employee['salary']}")
    print(f"Gender: {employee['gender']}")
    print(f"Employee_Level: {employee['employee_level']}")
    print(f" ")
    payslip_count+=1
    print(f"total payslips generated:{payslip_count}")
```
